```python
# intersted cols:
columns = ["SalePrice","OverallQual","GrLivArea",
           "GarageCars","FullBath","YearBuilt",
           "BedroomAbvGr","BldgType_1Fam",'BldgType_2fmCon', 'BldgType_Duplex', 'BldgType_Twnhs',
       'BldgType_TwnhsE', 'CentralAir_N', 'CentralAir_Y', 'HouseStyle_1.5Fin',
       'HouseStyle_1.5Unf', 'HouseStyle_1Story', 'HouseStyle_2.5Fin',
       'HouseStyle_2.5Unf', 'HouseStyle_2Story', 'HouseStyle_SFoyer',
       'HouseStyle_SLvl']
df = dummy_v[columns]
df.head()
```

# Linear Regression Model

```python
from sklearn.cross_validation import train_test_split
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
X = df.drop(["SalePrice"],axis = 1)
y = df["SalePrice"].reshape(-1,1)
```

```python
# fit model and R2 score
x_train,x_test,y_train,y_test=train_test_split(X , y , test_size = 0.3,random_state =42)
model.fit(x_train, y_train)
model.score(x_test,y_test)
```

# Gradient Boosting Model

```python
from sklearn import ensemble
clf = ensemble.GradientBoostingRegressor(n_estimators = 330, max_depth = 4, min_samples_split = 2,learning_rate = 0.1, loss = "ls")
```

```python
# fit the model
clf.fit(x_train, y_train)
```

```python
# print R2 score
acc_clf = clf.score(x_test,y_test)
acc_clf
```

```python
clf_pred = clf.predict(x_test)
```

```python
price = clf_pred
print(price[:3])
print(y[:3])
```

# Lasso Ridge ElasticNet Models

```python
# linear model 
# fit really well, to predict training set and test set, Biased variance tradeoff (predict training data well, but not the test data)
# regularization: avoid biased predict training data only. (not too much variance, and biased)
# Lasso = L1 algorithms, Ridge = L2 lgorithms, ElasticNet just the balance bwtn the 2
from sklearn.linear_model import Lasso, Ridge, ElasticNet
from sklearn.metrics import r2_score, mean_squared_error

```

```python
# lasso regression
# declare (finding the algorithms)
lasso = Lasso(random_state = 42)
# fit data
lasso_model = lasso.fit(x_train, y_train)
# prediction
y_predlasso = lasso_model.predict(x_test)
RMSE_lasso = mean_squared_error(y_test,y_predlasso)**0.5 #smaller scale, easier to look at (root mean sqaured error)
r2_lasso = r2_score(y_test,y_predlasso)
print("Lasso RMSE: ",RMSE_lasso)
print("Lasso R2: ",r2_lasso)
# need to do training as well, if r2 and RMSE similar, means it's not over fit
```

```python
ridge = Ridge(random_state = 42)
# fit data
ridge_model = ridge.fit(x_train, y_train)
# prediction
y_predridge = ridge_model.predict(x_test)
RMSE_ridge = mean_squared_error(y_test,y_predridge)**0.5 #smaller scale, easier to look at (root mean sqaured error)
r2_ridge = r2_score(y_test,y_predridge)
print("Ridge RMSE: ",RMSE_ridge)
print("Ridge R2: ",r2_ridge)
```

```python
elas = ElasticNet(random_state = 42)
# fit data
elas_model = elas.fit(x_train, y_train)
# prediction
y_predelas = elas_model.predict(x_test)
RMSE_elas = mean_squared_error(y_test,y_predelas)**0.5 #smaller scale, easier to look at (root mean sqaured error)
r2_elas = r2_score(y_test,y_predelas)
print("ElasticNet RMSE: ",RMSE_elas)
print("ElasticNet R2: ",r2_elas)
```
